```python
# criando um objeto com a base.  obs.: todo vez que inicializar o notebook no colaba será necessário carregar o arquivo.

filmes = pd.read_csv('/content/filmes_marvel.csv')

filmes.head(3)
```

```python
# criando o objeto eda => "exploratory data analysis"
# fç analyze() >> usada para um único dataset a ser visualizado, e não há necessidade comparar subpopulações (ex.: homens e mulheres)

eda = sweetviz.analyze(filmes)  
```

Usando a biblioteca Sweetviz, verifique qual foi o valor de orçamento (budget) que aconteceu mais vezes nos filmes da Marvel, e indique qual a porcentagem de vezes que esse valor foi determinado para os filmes.

Resposta.: o orçamento de 200 mi foi o que apareceu em 17% dos filmes. 

```python
# Conhecendo o "sweetviz" (Fast & powerful EDA in notebooks using Sweetviz 2.0), disponível em:
# https://colab.research.google.com/drive/1-md6YEwcVGWVnQWTBirQSYQYgdNoeSWg?usp%3Dsharing#scrollTo=C-oQkEXAPMsd

# método "show_notebook" nos traz a visualização dos dados no pp notebook do colab, como a seguir. 
eda.show_notebook(w='100%', h = 700) #(w=1200, h=1200, scale=1.0) estes parametros podem ser alterados para ajustar a exibição. 

# a visualização pode ser também em html:
# eda.show_html()
```

```python
# Usando uma correlação de Pearson, verifiquei no database da Marvel, 
# quais são as variáveis que tem maior correlação positiva?

mask = np.triu(np.ones_like(filmes.corr(), dtype=np.bool))

sns.heatmap(filmes.corr(),mask = mask, square = True, annot=True, vmin=-1, vmax=1)
plt.show()

# Resposta: Indicação para o Oscar e vencedor do oscar (coef corr = 0.9, ou seja, há uma gde chance de um 
# filme receber/vencer o Oscar qdo ele é indicado) e metascore e indicação p/ o oscar (0.63)
```
